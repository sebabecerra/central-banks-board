# Demo: usar `search_google.py` desde Python

Este notebook muestra la forma mas simple de importar y ejecutar la funcion `search_google`.

```python
import sys
from pathlib import Path

project_path = Path('/Users/sbc/projects/central-banks-board')
if str(project_path) not in sys.path:
    sys.path.append(str(project_path))

print(project_path)
```

## 1. Importar la funcion

```python
from search_google import search_google
```

## 2. Hacer una busqueda

La funcion recibe:
- `query`: texto que quieres buscar en Google
- `limit`: cuantos resultados quieres
- `language`: idioma de Google
- `country`: pais para Google
- `pause_seconds`: pausa entre paginas

```python
results = search_google(
    'Central Bank of Chile board',
    limit=5,
    language='en',
    country='us',
    pause_seconds=1.0,
)

len(results)
```

## 3. Ver los resultados

```python
for r in results:
    print('TITLE:', r.title)
    print('URL:', r.url)
    print('SNIPPET:', r.snippet)
    print('-' * 80)
```

## 4. Pasarlo a tabla

```python
import pandas as pd

df = pd.DataFrame([
    {
        'position': r.position,
        'title': r.title,
        'url': r.url,
        'snippet': r.snippet,
    }
    for r in results
])

df
```

## 5. Buscar en un sitio especifico

Si quieres restringir la busqueda a un dominio, solo agregalo dentro del query.

```python
results_site = search_google(
    'site:bcentral.cl board',
    limit=5,
    language='en',
    country='us',
    pause_seconds=1.0,
)

pd.DataFrame([
    {
        'position': r.position,
        'title': r.title,
        'url': r.url,
        'snippet': r.snippet,
    }
    for r in results_site
])
```
